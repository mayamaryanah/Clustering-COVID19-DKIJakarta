## 2. Data Preprocessing

```python
dataset.isna().sum()
```

```python
dataset['BELUM VAKSIN'].fillna(dataset['BELUM VAKSIN'].median(), inplace=True)
dataset['JUMLAH\nDOSIS 1'].fillna(dataset['JUMLAH\nDOSIS 1'].median(), inplace=True)
dataset['JUMLAH\nDOSIS 2'].fillna(dataset['JUMLAH\nDOSIS 2'].median(), inplace=True)
dataset['JUMLAH\nDOSIS 3'].fillna(dataset['JUMLAH\nDOSIS 3'].median(), inplace=True)
```

```python
dataset.duplicated().sum()
```

```python
import scipy.stats as stats
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(dataset)}')

filtered_entries = np.array([True] * len(dataset))

for col in nums:
    zscore = abs(stats.zscore(dataset[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
dataset = dataset[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 3

print(f'Jumlah baris setelah memfilter outlier: {len(dataset)}')
```

```python
datascaled=dataset[['POSITIF','Dirawat','Sembuh','Meninggal.1','Self Isolation','BELUM VAKSIN','JUMLAH\nDOSIS 1','JUMLAH\nDOSIS 2','JUMLAH\nDOSIS 3']]
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
kolom_all = [x for x in datascaled.columns]
for kolom in kolom_all:
  datascaled[kolom] = scale.fit_transform(np.array(datascaled[kolom]).reshape(-1,1))
```

## 3. Model and Evaluation

```python
from sklearn.cluster import KMeans


inertia = []

for i in range(2, 16):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

print(inertia)
plt.figure(figsize=(20, 10))
plt.title('Elbow Method of K-means Clustering')
# plt.plot(inertia)

sns.lineplot(x=range(2, 16), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 16), y=inertia, s=300, color='#800000',  linestyle='--')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 16)
eval_scores = []
for k in K:
    clusterer = KMeans(n_clusters=k, random_state=0)
    preds = clusterer.fit_predict(data)

    score = silhouette_score(data, preds)
    eval_scores.append(score)
    print("For n_clusters = {}, silhouette score is {}".format(k, score))
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(data)
data['Cluster'] = kmeans.labels_
dataset['Cluster'] = kmeans.labels_
```

```python
dataset
```

```python
arr_silhouette_score_euclidean = []
for i in K:
    kmeans = KMeans(n_clusters=i).fit(data)
    preds = kmeans.predict(data)
    
    score_euclidean = silhouette_score(data, preds, metric='euclidean')
    arr_silhouette_score_euclidean.append(score_euclidean)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x=range(2,16), y=arr_silhouette_score_euclidean, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2,16), y=arr_silhouette_score_euclidean, s=300, color='#800000',  linestyle='--')
```

## 4. Statistics Inferensia

```python
import scipy.stats as st
```

```python
# Langkah 1
# H0 : Rata-rata POSITIF sama untuk semua cluster
# H1 : Rata-rata POSITIF berbeda untuk beberapa cluster

# Langkah 2
df_Cluster0 = dataset[dataset['Cluster'] == 0]
df_Cluster1 = dataset[dataset['Cluster'] == 1]
df_Cluster2 = dataset[dataset['Cluster'] == 2]
df_Cluster3 = dataset[dataset['Cluster'] == 3]
```

```python
anova = st.f_oneway(df_Cluster0['POSITIF'],
                    df_Cluster1['POSITIF'],
                    df_Cluster2['POSITIF'],
                    df_Cluster3['POSITIF'])
```

```python
anova
```

```python
pvalue = anova.pvalue
print(pvalue)
if pvalue >= 0.05:
  print('Insufficient Reject H0')
else:
  print('Sufficient Reject H0')
```
